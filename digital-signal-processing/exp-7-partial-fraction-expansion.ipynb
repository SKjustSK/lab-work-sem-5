# 7) Compute partial fraction expansion of rational z-transform using PYTHON.

### 11 Nov 2025

```python
import sympy as sp
from IPython.display import display

z = sp.symbols('z')
numerator = z**2 + 1
denominator = 3*z**2 + 4*z + 2

z_transform = numerator / denominator
partial_fraction = sp.apart(z_transform,  full=True).doit()

print("Original: ")
display(z_transform)

print("Partial Fraction: ")
display(partial_fraction)

n1 = z**3 + 3*z**2 + 5*z + 4
d1 = 5*z**4 + 6*z**3 + 7*z**2 + 3*z + 5
z1 = n1 / d1
p1 = sp.apart(z1, full=True).doit()
display(z1)
display(p1)
```
